```python
var = pca.explained_variance_ratio_
```

```python
plt.figure()
sns.set_style('darkgrid')
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.show()
```

```python
a=np.cumsum(var)
a=np.round(a,2)
plt.bar(np.arange(len(a)),a)
plt.yticks(np.arange(0, 1, step=0.05))
plt.xlabel(" principal components")
plt.ylabel("variance cumulative")
plt.show()
```

```python
pca = PCA(n_components=2)
```

```python
pca_values=pca.fit_transform(x)
```

```python
pca.components_
```

```python
pca_values = pd.DataFrame(data = pca_values
             , columns = ['pc1', 'pc2'])
```

```python
pca_values
```

```python
final_df = pd.concat([pca_values, iris[['Species']]], axis = 1)
```

```python
final_df
```

```python
sns.scatterplot(x="pc1", y="pc2", hue="Species",data=final_df)
```

```python
from sklearn.cluster import KMeans

k_rng=range(1,13)
sse=[] # SUM OF SQUARED ERRORS
for k in k_rng:
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(pca_values)
    sse.append(kmeans.inertia_)
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_rng,sse)
from kneed import KneeLocator
kn = KneeLocator(k_rng, sse, curve='convex', direction='decreasing')
print(kn.knee)
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(pca_values)
labels=kmeans.labels_ 
labels
centers=kmeans.cluster_centers_
centers
```

```python
pca_values['cluster']=labels
pca_values1=pca_values[pca_values.cluster==0]
pca_values2=pca_values[pca_values.cluster==1]
pca_values3=pca_values[pca_values.cluster==2]
```
