```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
stock=pd.read_csv("c:\\Users\\kumar\\EW-MAX.csv")
```

```python
stock
```

### HISTOGRAM

```python
for c in stock.columns:
    if c == "Date": 
        continue
    else:
        sns.FacetGrid(stock,height=5)        .map(sns.distplot,c)        .add_legend()
        plt.show()
```

```python
for c in stock.columns:
    if c == "Date":
        continue
    else:
        plt.hist(stock[c])
        plt.xlabel(c)
        plt.show()
```

### BOXPLOT

```python
for c in stock.columns:
    if c == "Date":
        continue
    else:
        plt.boxplot(stock[c])
        plt.xlabel(c)
        plt.show()
```

### PAIRPLOT(SCATTER PLOTS COMBINED)

```python
sns.pairplot(stock)
```

### LINE PLOT

```python
sns.lineplot(data=stock.drop(['Date'], axis=1))
```

### HEAT MAP

```python

corr = stock.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
```

```python
sns.heatmap(stock.corr(), annot=True)
```
